```python
import pandas_datareader.data as web
import statsmodels.api as sm
import matplotlib
```

```python
data_web = web.DataReader(['GE', 'MSFT'], 'yahoo', start='2019-01-01', end='2020-01-01')
```

```python
data_web['High']['MSFT'].plot();
```

```python
from matplotlib import pyplot
decompose = sm.tsa.seasonal_decompose(data_web['High']['MSFT'], freq=30)
```

```python
decompose.seasonal.plot();
```

```python
decompose.trend.plot();
```

```python
decompose.resid.plot();
```

```python
from matplotlib.pyplot import figure

figure(num=1, figsize=(20, 20), dpi=80)

decompose.plot()
```
